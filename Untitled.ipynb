```python
# Dependencies
import pandas as pd
import glob
```

```python
#Merging csv files (without repeating header)
January_March2019_files = glob.glob("*.csv") #(thanks to Josh who slaked this funtion)
df = pd.concat((pd.read_csv(f, header = 0) for f in January_March2019_files))
df
```

```python
#replace gender
df["Gender"]= df['gender'].replace([0,1,2],["unknown", "male", "female"])
df=df.drop('gender',1)
df.head()
```
